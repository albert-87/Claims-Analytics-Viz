# 3. Data Wrangling

```python
# delete unwanted columns
df = df.drop(['max_torque', 'max_power','engine_type','is_esc','is_adjustable_steering','is_tpms','rear_brakes_type','is_front_fog_lights','is_rear_window_wiper','is_rear_window_washer','is_rear_window_defogger','gear_box','displacement','policy_tenure','area_cluster','is_parking_sensors','is_power_door_locks','is_central_locking','is_power_steering','is_driver_seat_height_adjustable','is_day_night_rear_view_mirror','is_ecw','is_speed_alert','ncap_rating', 'airbags', 'cylinder','segment', 'fuel_type','make'], axis=1)
```

```python
df.head()
```

```python
# Convert age columns to real age
```

- age_of_car 
- age_of_policyholder need to be converted to real age


```python
# convert age_of_policyholder values from 0-1 to 0-70
df['age_of_policyholder'] = df['age_of_policyholder'] * 70

# round off the values to the nearest integer
df['age_of_policyholder'] = df['age_of_policyholder'].round(3)

# find the smallest and largest value in the column
smallest_age = df['age_of_policyholder'].min()
largest_age = df['age_of_policyholder'].max()

# print the results
print("Smallest age in the column is:", smallest_age)
print("Largest age in the column is:", largest_age)
```

```python
# convert age_of_car values 
df['age_of_car'] = df['age_of_car'] * 70

# round off the values to the nearest integer
df['age_of_car'] = df['age_of_car'].round(5)

# find the smallest and largest value in the column
smallest_age = df['age_of_car'].min()
largest_age = df['age_of_car'].max()

# print the results
print("Smallest age in the column is:", smallest_age)
print("Largest age in the column is:", largest_age)
```

```python
# define the car models and makes
car_models = ['Ford Fiesta', 'Volkswagen Golf', 'Vauxhall Corsa', 'Ford Focus', 'Mercedes-Benz A-Class', 'Nissan Qashqai', 'Mini', 'Volkswagen Polo', 'Kia Sportage', 'Audi A3', 'BMW 3 Series']

# create a dictionary mapping car models to "M" values
model_values = {'M1': car_models[0], 'M2': car_models[1], 'M3': car_models[2], 'M4': car_models[3],
                'M5': car_models[4], 'M6': car_models[5], 'M7': car_models[6], 'M8': car_models[7],
                'M9': car_models[8], 'M10': car_models[9], 'M11': car_models[10]}

# print the updated dictionaries
print(model_values)
```

```python
# use the map function to replace values in the corresponding columns
df['model_values'] = df['model'].map(model_values)
df.head()
```
